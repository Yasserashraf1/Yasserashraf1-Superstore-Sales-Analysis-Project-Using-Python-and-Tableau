2. Exploratory Data Analysis (EDA):

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')
```

```python
df[['Order Date','Ship Date']]
```

```python
df = df.drop('Row ID', axis=1)
```

```python
df
```

```python
df['Ship Mode'].value_counts()
```

```python
labels = df['Ship Mode'].value_counts().index
sizes = df['Ship Mode'].value_counts()
colors = ['#ff9999'] * len(sizes)
colors[0] = '#66b3ff'
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title('Ship Mode Distribution')
plt.show()
```

```python
df['Segment'].value_counts()
```

```python
labels = df['Segment'].value_counts().index
sizes = df['Segment'].value_counts()
colors = ['#ff9999'] * len(sizes)
colors[0] = '#66b3ff'
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title('Segment Distribution')
plt.show()
```

```python
numerical_columns = df.select_dtypes(include=['int','float64']).columns

plt.figure(figsize=(20, 8))
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2,4 , i)
    df.boxplot(column=col)
    plt.title(col)
    plt.tight_layout()

plt.show()
```

```python
numerical_columns = df.select_dtypes(include=['int','float64']).columns

plt.figure(figsize=(20, 8))
colors = ['blue', 'red']
for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 4, i)
    df[col].hist(color=colors[i-1])
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()

plt.show()
```
